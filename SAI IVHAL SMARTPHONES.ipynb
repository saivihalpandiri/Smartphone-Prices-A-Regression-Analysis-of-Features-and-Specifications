## REGRESSION MODEL

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,BaggingRegressor,AdaBoostRegressor
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split,cross_val_score,KFold,StratifiedKFold,ShuffleSplit,LeaveOneOut,GridSearchCV
from sklearn.metrics import r2_score
```

```python
x=df.drop(columns=['price'])
y=df['price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=10,test_size=0.2)
```

```python
model=[DecisionTreeRegressor(criterion="squared_error",min_samples_split=2,max_features=0.8,max_depth=30)
       ,RandomForestRegressor(n_estimators=80,min_samples_split=2,max_features=0.8,max_depth=30,max_samples=0.6),
       BaggingRegressor(n_estimators=100,max_features=0.8,max_samples=0.6),
       SVR(kernel="linear",gamma=0.01,C=10),
       KNeighborsClassifier(n_neighbors=5)]
```

```python
for i in model:
    i.fit(x_train,y_train)
    Pred=i.predict(x_test)
    print(r2_score(y_test,Pred))
```

## GRID SEARCH

```python
grid_param1={
    "criterion":["squared_error","gini"],
    "min_samples_split":[1,2,3,4,5,6],
    "max_features":[1,0.9,0.83,0.7,0.6],
    "max_depth":[20,30,40,50,60]
}
```

```python
grid_param2={
    "n_estimators":[50,70,80,900],
    "min_samples_split":[2,3,4.7,5,6],
    "max_features":[1,0.9,0.8,0.7,0.6],
    "max_depth":[20,30,40,50],
    "max_samples":[1,0.9,0.8,0.7,0.6]
}
```

```python
grid_param3={
    "n_estimators":[70,80,90,100],
    "max_features":[1,0.9,0.8,0.7,0.6],
    "max_samples":[1,0.9,0.8,0.7,0.6]
}
grid_param4={
    "kernel":["linear","rbf"],
    "gamma":[0.1,0.01],
    "C":[100,80]
}
```

```python
gscv=GridSearchCV(estimator=DecisionTreeRegressor(criterion="squared_error",min_samples_split=2,max_features=0.8,max_depth=30),param_grid=grid_param1,n_jobs=-1,verbose=2)
gscv.fit(x,y)
gscv.best_params_
```

```python
gscv=GridSearchCV(estimator=RandomForestRegressor(n_estimators=80,min_samples_split=2,max_features=0.8,max_depth=30,max_samples=0.6),param_grid=grid_param2,n_jobs=-1,verbose=2)
gscv.fit(x,y)
gscv.best_params_
```

```python
gscv=GridSearchCV(estimator=BaggingRegressor(n_estimators=100,max_features=0.8,max_samples=0.6),param_grid=grid_param3,n_jobs=-1,verbose=2)
gscv.fit(x,y)
gscv.best_params_
```

```python
gscv=GridSearchCV(estimator=SVR(kernel="linear",gamma=0.01,C=100),param_grid=grid_param4,n_jobs=-1,verbose=2)
gscv.fit(x,y)
gscv.best_params_
```
